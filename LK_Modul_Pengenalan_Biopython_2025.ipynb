###**1.3. Penggabungan Sekuens**

Anda diminta untuk menggabungkan sekuens-sekuens berikut ini dengan `spacer ("GGGGGAA")` <br>
`seq_KimProt` = ("ATGCGATGCGTAAACGTGTCGAGC") <br>
`seq_ClarProt` = ("GTAACCCGGTGATCGACC") <br>
`seq_CarinProt` = ("GTCGAAACGATCGATCGTG") <br>
Gunakan [tutorial Biopython](https://biopython.org/docs/latest/Tutorial/index.html) untuk membantu mengerjakan soal ini!

**Contoh script**
    
    spacer = Seq("isi_spacer")    #Ganti isi_spacer sesuai yang diminta pada soal
    seq_KimProt = Seq("isi_sekuens")    #Ganti isi_sekuens dengan sekuens yang sesuai
    seq_ClarProt = Seq("isi_sekuens")
    seq_CarinProt = Seq("isi_sekuens")
    contigs = [seq_KimProt, ..., ...]    #Isi titik-titik dengan seq_ClarProt dan seq_CarinProt
    assembly = spacer.join(contigs)
    print(assembly)

Jawab :

````python
spacer = Seq("GGGGGAA")    #Ganti isi_spacer sesuai yang diminta pada soal
seq_KimProt = Seq("ATGCGATGCGTAAACGTGTCGAGC")    #Ganti isi_sekuens dengan sekuens yang sesuai
seq_ClarProt = Seq("GTAACCCGGTGATCGACC")
seq_CarinProt = Seq("GTCGAAACGATCGATCGTG")
contigs = [seq_KimProt, seq_ClarProt, seq_CarinProt]    #Isi titik-titik dengan seq_ClarProt dan seq_CarinProt
assembly = spacer.join(contigs)
print(assembly)
````

### 4.6. Parsing File .fastq dan Pembuatan Dataframe

Setelah melihat ketiga faktor identifikasi data metagenomik tersebut. Anda dapat menyusun data sebelum dianalis lebih lanjut, yaitu dengan *parsing* dan pembuatan *dataframe*. Anda diminta untuk parsing terhadap file .fastq dan pembuatan dataframe sebelum dilakukan analisis visualisasi. <br> Gunakan [tutorial Biopython](https://https://biopython.org/docs/latest/Tutorial/index.html) dan pandas untuk membantu mengerjakan soal ini!

Contoh Script

```
from Bio import SeqIO

# Parsing file FASTQ
filtered_sequences = []    #Ganti filtered_sequences dengan volcano_filtered dan format_file dengan format file yang sesuai
for record in SeqIO.parse(file_path, "format_file"):   
    filtered_sequences.append({
        "sequence": str(record.seq),
        "length": len(record.seq),
        "gc_content": (str(record.seq).count("G") + str(record.seq).count("C")) / len(record.seq) * 100,
        "avg_quality": sum(record.letter_annotations["phred_quality"]) / len(record.letter_annotations["phred_quality"])
    })

# Konversi ke DataFrame
import pandas as pd
df = pd.DataFrame(filtered_sequences) #Ganti df dengan volcano_df dan filtered_sequences dengan volcano_filtered
print(df.head())
```



Jawab :

````python
# Parsing file FASTQ
volcano_filtered = []    #Ganti filtered_sequences dengan volcano_filtered dan format_file dengan format file yang sesuai
for record in SeqIO.parse(file_path, "fastq"):
    volcano_filtered.append({
        "sequence": str(record.seq),
        "length": len(record.seq),
        "gc_content": (str(record.seq).count("G") + str(record.seq).count("C")) / len(record.seq) * 100,
        "avg_quality": sum(record.letter_annotations["phred_quality"]) / len(record.letter_annotations["phred_quality"])
    })

# Konversi ke DataFrame
import pandas as pd
volcano_df = pd.DataFrame(volcano_filtered) #Ganti df dengan volcano_df dan filtered_sequences dengan volcano_filtered
volcano_df.head()
````

### **4.7. Reduksi Dimensi PCA**

Anda diminta untuk melakukan reduksi dimensi dataframe sebelum dilakukan analisis visualisasi dengan *principal component analysis*. <br> Gunakan [tutorial Biopython](https://https://biopython.org/docs/latest/Tutorial/index.html) untuk membantu mengerjakan soal ini!

Contoh Script

```
from sklearn.decomposition import PCA

# Ekstraksi fitur
features = df[["length", "gc_content", "avg_quality"]].values   #Ganti df dengan volcano_df

principal_components = pca.fit_transform(features)

# Menyimpan hasil ke DataFrame
df_pca = pd.DataFrame(principal_components, columns=["PC1", "PC2"])   #Ganti df_pca dengan volcano_pca
```
