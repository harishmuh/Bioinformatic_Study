```python
# Exercise 1
# GC Content Calculator

# Pseudocode
# Mulai
# Input Sequence
# Menghitung panjang sequence
# Hitung jumlah G
# Hitung jumlah C
# Hitung jumlah A
# Hitung jumlah T
# Hitung GC content (G+C)/total, bentuk persen
# print sesuai format

# Input sequence
dna = "AGGCTAGCATTAGGCATCGGATTCGAGAG"

# Menghitung panjang sequence
# print("Length: ", len(dna))
panjang = len(dna)

# Hitung jumlah basa
jumlah_g = dna.count("G")
jumlah_c = dna.count("C")
jumlah_a = dna.count("A")
jumlah_t = dna.count("T")

#print("G count: ", dna.count("G"))
#print("C count: ", dna.count("C"))
#print("A count: ", dna.count("A"))
#print("T count: ", dna.count("T"))

# Hitung GC content
gc_content = (dna.count("G") + dna.count("C"))/len(dna)*100
#print("GC content: ", gc_content)

# print sesuai format
print("=== DNA Analysis Report ===")
print(f"Sequence:    {dna}")
print("Sequence:    ", dna)
print(f"Length:      {panjang} bp")
print(f"A count:     {jumlah_a}")
print(f"T count:     {jumlah_t}")
print(f"G count:     {jumlah_g}")
print(f"C count:     {jumlah_c}")
print(f"GC Content:  {gc_content:.1f}%")



```

```python
import re
import csv

# Step 1
filename = input("Masukkan nama file FASTA: ")

# Step 2
with open(filename, "r") as file:
    content = file.read()

# Step 3
entries = re.findall(r">(.*?)\n([A-Za-z\n]+)", content)

# Step 4
data = []

for name, sequence in entries:
    sequence = sequence.replace("\n", "")
    length = len(sequence)
    data.append([name, sequence, length])

# Step 5
with open("output.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Organism", "Sequence", "Length"])
    writer.writerows(data)

print("Data berhasil disimpan ke output.csv")
```
